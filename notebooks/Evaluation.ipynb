```python
def bug_columns(df, label='induces'):
    """return all columns from the bug-matrix"""
    jlip = []
    for col in df.columns:
        if col.startswith('{}__'.format(label)):
            jlip.append(col)
    return jlip

def load_project(project_name):
    """load project from the supplied csv."""
    df =  pd.read_csv('{}/{}.csv'.format(DATA_PATH, project_name))
    df['project'] = project_name
    df['committer_date'] = pd.to_datetime(df['committer_date'])
    return df

def last_commits(df, num_commits=500):
    """return last num_commits"""
    last_commits = []
    for c in df['commit'].unique()[-num_commits:]:  # order is preserved here (nice!)
        last_commits.append(c)
    return last_commits

def bugs_later_than(df, cutoff_date):
    """return columns from bug-matrix that are too late"""
    if not cutoff_date:
        raise Exception('please supply a cutoff date')

    remove = []
    for col in df.columns:
        if col.startswith('induces__'):
            label, issue_id, bugfix_commit, bugfix_date = col.split('__')
            if pd.to_datetime(bugfix_date, utc=True) > cutoff_date:
                remove.append(col)
    return remove

def score(train, test, features, use_smote=False):
    """simple scorer for demonstration"""
    if use_smote:
        sm = SMOTE(random_state=RANDOM_SEED)
        X_res, y_res = sm.fit_resample(train[features].values, train['is_inducing'].values)

    y_train = train['is_inducing']
    y_test = test['is_inducing']

    X_train = train[features]
    X_test = test[features]

    # trivial random forest
    rf = RandomForestClassifier()
    if use_smote:
        rf.fit(X_res, y_res)
    else:
        rf.fit(X_train, y_train)

    y_pred = rf.predict(X_test)
    mcc = matthews_corrcoef(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print('train instances: {}, true positive: {}'.format(len(X_train), sum(y_train)))
    print('test instances: {}, true positive: {}'.format(len(X_test), sum(y_test)))
    print('mcc', mcc)
    print('prec', prec)
    print('recall', recall)
    print('auc', auc)
    print('f1', f1)
```

```python
# load one project
df = load_project('jspwiki')
```

```python
# last 500 commits are test data
lc = last_commits(df, num_commits=500)

test = df[df['commit'].isin(lc)].copy()
train = df[~df['commit'].isin(lc)].copy()

# only count inducings that were fixed prior to the first of the 500 commits
late_bugs = bugs_later_than(df, cutoff_date=test['committer_date'].min())
train.drop(columns=late_bugs, inplace=True)

# add labels for simpler training
train_bugs = bug_columns(train)
train['is_inducing'] = train[train_bugs].any(axis=1)

test_bugs = bug_columns(test)
test['is_inducing'] = test[test_bugs].any(axis=1)

# we can now also apply a gap to remove data from the end of the training which may not be likely to have inducing labels
train = train.drop(train[train['committer_date'] > test['committer_date'].min() - relativedelta(months=3)].index)
```
