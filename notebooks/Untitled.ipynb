```python
import pandas as pd
import csv
import collections
from joblib import load
```

```python
with open('../scores/baseline_rf.csv', mode='r') as infile:
    reader = csv.reader(infile)
    rows = []
    for row in reader:
        rows.append(row)
```

```python
keys = rows.pop(0)
```

```python
bl_rf = {}
pr_names = [itm[0] for itm in rows]
keys.pop(0)
```

```python
keys
```

```python
counter = 0
for row in rows:
    name = pr_names[counter]
#         print(row[1], row[2], row[3], row[4], row[5])
    bl_rf[name] = {keys[0]: row[1],
                   keys[1]: row[2],
                   keys[2]: row[3],
                   keys[3]: row[4],
                   keys[4]: row[5]}
    counter += 1
```

```python
bl_rf
```

```python
result = collections.OrderedDict(sorted(bl_rf.items(), key=lambda t:t[1]["mcc"], reverse=True))
```

```python
bl_rf = dict(result)
```

```python
bl_rf
```

```python
pd.DataFrame(bl_rf).transpose()
```
